### 1. Scatterplot (Numerical - Numerical)

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))

sns.despine(f, left=False, bottom=False)

sns.scatterplot(x=tips['total_bill'], 
                y=tips['tip'], 
                hue=tips['sex'],
                style=tips['smoker'],
                size=tips['size'],
                palette="ch:r=-.2,d=.3_r",
                data=tips, ax=ax)
```

### 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.set_theme(style="whitegrid")

g = sns.barplot(data = titanic,
            x = titanic['Pclass'],
            y = titanic['Age'],
            hue=titanic['Embarked'],
            errorbar="sd",
            palette="dark", 
            alpha=.6)
```

### 3. Box Plot (Numerical - Categorical)

```python
sns.set_theme(style="ticks", palette="pastel")

sns.boxplot(x = titanic['Sex'],
            y = titanic['Age'],
            hue=titanic['Survived'],
            palette=["m", "g"])
```

### 4. Distplot (Numerical - Categorical)

```python
# seaborn.displot(data=None, *, x=None, y=None, hue=None, row=None, col=None, weights=None, kind='hist', rug=False, 
# rug_kws=None, log_scale=None, legend=True, palette=None, hue_order=None, hue_norm=None, color=None, col_wrap=None, 
# row_order=None, col_order=None, height=5, aspect=1, facet_kws=None, **kwargs)

sns.displot(titanic[titanic['Survived']==0]['Age'],kde=True)
sns.displot(titanic[titanic['Survived']==1]['Age'],kde=True)
```

### 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Embarked')['Survived'].mean() * 100).round(2)
```

### 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
