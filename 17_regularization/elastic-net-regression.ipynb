# Elastic Net Regularization

Elastic Net Regularization is a regression technique that combines the strengths of Lasso (L1) and Ridge (L2) regularization methods. It is particularly useful for handling datasets with high-dimensional features and multicollinearity (highly correlated features). By blending L1 and L2 penalties, Elastic Net provides a balance between feature selection and coefficient shrinkage, making it a versatile tool for predictive modeling.

!

### Key Concepts
Elastic Net regularization addresses the limitations of Lasso and Ridge regression. While Lasso can drive some coefficients to zero for feature selection, it struggles with correlated features, often selecting only one feature from a group. Ridge, on the other hand, shrinks coefficients but does not perform feature selection. Elastic Net combines these penalties using two hyperparameters:

**Alpha ()**: Controls the overall strength of regularization.<br>
**L1 Ratio ()**: Determines the mix of L1 and L2 penalties. A value of 0 corresponds to Ridge regression, 1 to Lasso, and values between 0 and 1 to a mix of both


### Advantages and Applications
- Feature Selection: Elastic Net can drive irrelevant feature coefficients to zero, simplifying models and improving interpretability
- Handling Multicollinearity: Unlike Lasso, Elastic Net selects all correlated features together, making it ideal for datasets with multicollinearity
- High-Dimensional Data: It performs well with datasets containing many features, especially when the number of features exceeds the number of samples
- Regression Tasks: Elastic Net is widely used in linear regression models to predict continuous target variables

### Use Cases:
- Genomics: Selecting relevant genes in high-dimensional datasets
- Finance: Portfolio optimization by identifying influential factors
- Machine Learning: Improving model generalization and reducing overfitting

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Linear Regression
reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# Ridge 
reg = Ridge(alpha=0.1)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# Lasso
reg = Lasso(alpha=0.01)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# ElasticNet
reg = ElasticNet(alpha=0.005,l1_ratio=0.9)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```
